```
#/FileStore/tables/Machine_learning/
from pyspark.sql import SparkSession
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.evaluation import BinaryClassificationEvaluator, MulticlassClassificationEvaluator

spark = SparkSession.builder.appName('mylogreg').getOrCreate()
```

```
my_data = spark.read.format('libsvm').load('/FileStore/tables/Machine_learning/sample_libsvm_data.txt')
```

```
my_data.show()
```

```
my_logreg_model = LogisticRegression()
```

```
fitted_logreg_model = my_logreg_model.fit(my_data)
```

```
log_summary = fitted_logreg_model.summary
```

```
#log_summary.predictions.printSchema()
log_summary.predictions.show()
```

```
train_set, test_set = my_data.randomSplit([0.7, 0.3])
```

```
final_model = LogisticRegression()
```

```
fit_final = final_model.fit(train_set)
```

```
prediction_lables = fit_final.evaluate(test_set)
```

```
prediction_lables.predictions.show()
```

```
my_eval = BinaryClassificationEvaluator()
```

```
my_final_roc = my_eval.evaluate(prediction_lables.predictions)
my_final_roc
```
