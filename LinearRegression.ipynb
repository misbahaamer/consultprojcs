```
from pyspark.sql import SparkSession
from pyspark.ml.regression import LinearRegression


spark = SparkSession.builder.appName('lrex').getOrCreate()
```

```
training = spark.read.format('libsvm').load("/FileStore/tables/Machine_learning/sample_linear_regression_data.txt")
```

```
training.show()
```

```
lr = LinearRegression(featuresCol = 'features', labelCol = 'label', predictionCol = 'prediction') #recall ML algorithm LinearRegression to define features and label columns and anso to declare predection col if u need to
lrModel = lr.fit(training) #fit the data into the ml model
```

```
print ("coefficients =", lrModel.coefficients) #coefficients and intercept of the lr model
print ("intercept =", lrModel.intercept)
```

```
training_summary = lrModel.summary
print ("R2 =", training_summary.r2)
print ("root mean squared error =", training_summary.rootMeanSquaredError)
```

```
all_data = spark.read.format('libsvm').load("/FileStore/tables/Machine_learning/sample_linear_regression_data.txt")
```

```
training_data, test_data = all_data.randomSplit([0.7, 0.3])
training_data.describe().show()
test_data.describe().show()
```

```
correct_model = lr.fit(training_data)
```

```
test_results = correct_model.evaluate(test_data)
print(test_results.residuals.show())
print ("R2 =", test_results.r2)
print ("root mean squared error =", test_results.rootMeanSquaredError)
```

```
unlabeled_data = test_data.select('features')
#unlabeled_data.show()
```

```
predictions = correct_model.transform(unlabeled_data)
predictions.show()
```
