```
from pyspark.sql import SparkSession
from pyspark.ml.clustering import KMeans
```

```
spark = SparkSession.builder.appName('kclusterringdocex').getOrCreate()
```

```
data = spark.read.format('libsvm').load('/FileStore/tables/Machine_learning/sample_kmeans_data.txt')
```

```
data.show()
```

```
final_data = data.select('features')
```

```
kmeans = KMeans().setK(3).setSeed(1)
```

```
model = kmeans.fit(final_data)
```

```
wssse = model.computeCost(final_data) #within sum of squraed errors
```

```
print(wssse)
```

```
centers = model.clusterCenters()
```

```
centers
```

```
results = model.transform(final_data)
```

```
results.show()
```
